```python
data = '''Несчастная кошка порезала лапу-
Сидит, и ни шагу не может ступить.
Скорей, чтобы вылечить кошкину лапу
Воздушные шарики надо купить!

И сразу столпился народ на дороге-
Шумит, и кричит, и на кошку глядит.
А кошка отчасти идет по дороге,
Отчасти по воздуху плавно летит!'''

print(data)
print()
print(len(data))

data1 = [word.strip('. , ! -').lower() for word in data.split()]
data1 = [word.lower() for word in data1]

print()
print(data1)
print()
print(len(data1))

dict1 = {}
for word in data1:
    if word not in dict1.keys():
        dict1[word] = 1
    else:
        dict1[word] += 1

total = 0
        
for i in dict1:
    if dict1[i] == 1:
        total += 1
print()
print(total)
print()

c = len(data1)

for i in range(c-1, c-11, -1):
    print(data1[i])

ans = "\n"
ans = ans.join(data1)

print()
print(ans)

print()
print(ans[19:30])

print()

ans2 = []

for i in range(len(data1)-1):
    ans2.append((data1[i], data1[i+1]))

print(ans2)
print()

dict2 = {}
for bigramm in ans2:
    if bigramm not in dict2.keys():
        dict2[bigramm] = 1
    else:
        dict2[bigramm] += 1

for bigramm in dict2:
    print("{} : {}".format(bigramm, dict2[bigramm]))
```
